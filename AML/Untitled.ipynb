```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the univariate normal distribution
mean = 0
std_dev = 1

# Generate data
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Plotting the normal distribution
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Normal Distribution')
plt.title('Univariate Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parameters for the bivariate normal distribution
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # Covariance matrix

# Generate grid and multivariate normal distribution
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))
rv = multivariate_normal(mean, cov)
z = rv.pdf(pos)

# Plotting the multivariate normal distribution
plt.figure(figsize=(8, 6))
plt.contourf(x, y, z, cmap='viridis')
plt.colorbar(label='Probability Density')
plt.title('Bivariate Normal Distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
